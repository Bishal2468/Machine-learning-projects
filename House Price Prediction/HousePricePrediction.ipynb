Import the Dependency

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRFRegressor
from sklearn import metrics
from sklearn.metrics import accuracy_score
import sklearn.datasets
```

```python
#iMPORTING the 
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
#Loading the dataset to pandas dataframe
house_price_dataframe = pd.DataFrame(housing.data, columns=housing.feature_names)
#print firstfive rows of the data frame
house_price_dataframe.head()
house_price_dataframe['target'] = housing.target
house_price_dataframe.head()
```

```python
house_price_dataframe.shape
```

```python
#checking the missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measures of the dataset
house_price_dataframe.describe()
```

Undestanding the correlation between various features in the dataset
1. postitive correlation
2. negative correlation

```python
correlation = house_price_dataframe.corr()
```

```python
#constructing a heatmap to understand correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f',annot = True,annot_kws={'size':8},cmap='Blues')
```

splitting the data and target

```python
X = house_price_dataframe.drop(['target'],axis=1)
Y = house_price_dataframe['target']
print(X)
print(Y)
```

```python
#Splitting the data into Training and and Test Data
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=3)
print(X.shape,X_train.shape,X_test.shape)
```

Model Training - XGBoost Regressor

```python
#Loading the model
model = XGBRFRegressor()
#training the model with X_train
model.fit(X_train,Y_train)
```

Evalution 

```python
#Prediction n the training data
#Accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
#R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)
#Mean absolute Error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print('R square error',score_1)
print('Mean absolute Error',score_2)
```

```python
#Prediction on test data
test_data_prediction = model.predict(X_test)
score_1 = metrics.r2_score(Y_test,test_data_prediction)
score_2= metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print('Test R square error : ',score_1)
print('Test mean square error :',score_2)
```

Visualizing the actual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual price')
plt.ylabel('Predicted  price')
plt.title('Actual Price vs Predicted price')
plt.show()
```
